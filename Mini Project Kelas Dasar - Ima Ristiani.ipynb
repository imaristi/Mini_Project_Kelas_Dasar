##Latar Belakang Dataset

Olimpiade Tokyo 2020 disebut sebagai Olimpiade Musim Panas 2020 disebut juga sebagai Olimpiade XXXII (Olimpiade ke-32) adalah acara multi-olahraga internasional yang diadakan pada tanggal 23 Juli hingga 8 Agustus 2021 di Tokyo, Jepang, dengan beberapa acara awal yang dimulai pada tanggal 21 Juli.

Tokyo terpilih sebagai kota tuan rumah selama Sesi IOC ke-125 di Buenos Aires, Argentina, pada 7 September 2013. Awalnya Olimpiade ini dijadwalkan berlangsung dari 24 Juli hingga 9 Agustus 2020, acara tersebut ditunda hingga 2021 pada Maret 2020 sebagai akibat dari pandemi COVID-19. Namun, acara tersebut mempertahankan nama Tokyo 2020 untuk tujuan pemasaran dan branding. Sebagian besar diadakan di balik pintu tertutup tanpa penonton publik yang diizinkan karena deklarasi keadaan darurat di Wilayah Tokyo Raya sebagai tanggapan terhadap pandemi. Paralimpiade Musim Panas akan diadakan antara 24 Agustus dan 5 September 2021, 16 hari setelah selesainya Olimpiade.

Olimpiade 2020 adalah Olimpiade keempat yang diadakan di Jepang, setelah Olimpiade Tokyo 1964 (Musim Panas), Sapporo 1972 (Musim Dingin) dan Nagano 1998 (Musim Dingin). Tokyo adalah kota pertama di Asia yang mengadakan Olimpiade Musim Panas dua kali. Olimpiade 2020 adalah yang kedua dari tiga Olimpiade berturut-turut yang akan diadakan di Asia Timur, setelah Olimpiade Musim Dingin 2018 di Pyeongchang, Korea Selatan dan sebelum Olimpiade Musim Dingin 2022 di Beijing, Cina.

Acara baru diperkenalkan dalam olahraga yang ada untuk tahun 2020, termasuk bola basket 3x3, BMX gaya bebas dan acara tim gender campuran di sejumlah olahraga yang ada, serta kembalinya bersepeda madison untuk pria dan pengenalan acara yang sama untuk wanita. Kebijakan IOC baru juga memungkinkan panitia penyelenggara tuan rumah untuk menambahkan olahraga baru ke program Olimpiade hanya untuk satu Pertandingan. Disiplin yang ditambahkan oleh Komite Olimpiade Jepang adalah bisbol dan softball, karate, olahraga panjat tebing, berselancar dan skateboard, empat terakhir yang membuat debut Olimpiade mereka.

Amerika Serikat menduduki puncak jumlah medali dengan total emas (39) dan total medali (113), dengan China menempati posisi kedua dengan kedua belah orang (38 dan 88). Negara tuan rumah Jepang menempati posisi ketiga, menetapkan rekor medali emas terbanyak dan total medali yang pernah dimenangkan oleh delegasi mereka di Olimpiade dengan 27 dan 58. Inggris menempati posisi keempat, dengan total 22 medali emas dan 65 medali, menjadi negara pertama di Olimpiade Musim Panas yang menambah atau menyamai total medali mereka yang dimenangkan dalam dua Pertandingan setelah menjadi tuan rumah mereka. Delegasi Rusia yang bersaing saat ROC menempati posisi kelima dengan 20 medali emas dan ketiga dalam jumlah medali keseluruhan, dengan 71 medali. Bermuda, Filipina dan Qatar memenangkan medali emas Olimpiade pertama mereka. Burkina Faso, San Marino dan Turkmenistan memenangkan medali Olimpiade pertama mereka.

## Data Visualization

###Dari negara mana saja sebagian besar & sebagian kecil atlet berasal?




```
y = athletes.NOC.value_counts().values[0:31]
x = athletes.NOC.value_counts().index[0:31]
plt.figure(figsize=(20,5))
plt.bar(x,y)
plt.xlabel("Country")
plt.ylabel("Number of athletes")
plt.xticks(rotation='vertical',size=10)
plt.yticks(size=15)
plt.show()
```

```
athletes.NOC.value_counts()
```

Dari hasil diagram diatas, sebagian besar atlet yang berpartisipasi dalam Olimpiade Tokyo 2020 ini berasal dari Amerika Serikat (615 atlet), Jepang (586 atlet), Australia (470 atlet), Cina (401 atlet) dan Jerman (400 atlet). Kemudian sebagian kecil atlet berasal dari negara-negara seperti Tanzania (2 atlet), Andorra (2 atlet), Myanmar (2 atlet), Sudan Selatan (2 atlet), Pulau Marshall (2 atlet) dan masih banyak lagi.

###"Discipline" mana yang paling populer dan negara mana yang memiliki peserta tertinggi di dalamnya?

```
y = athletes.Discipline.value_counts().values
x = athletes.Discipline.value_counts().index
plt.figure(figsize=(20,8))
plt.bar(x,y,width=0.5)
for index, value in enumerate(y):
    plt.text(index, value, str(value),color='blue',size=10,fontweight='bold')
plt.xlabel("Discipline")
plt.ylabel("Number of athletes")
plt.xticks(rotation='vertical')
plt.show()
```

Dari hasil diagram diatas, Atletik merupakan "Discipline" paling populer di Olimpiade Tokyo 2020. Atletik ini memiliki jumlah peserta yang paling banyak (2068 atlet). Berenang (743 atlet) dan Sepak Bola (567 atlet) adalah yang paling populer berikutnya. Kemudian, Bersepeda BMX Freestyle adalah "Discipline" yang paling tidak populer dengan hanya 19 atlet di dalamnya.

```
athletics_data = teams[(teams["Discipline"] == "Athletics")]
athletics_data["NOC"].value_counts()
```

Dari data diatas, negara yang sebagian besar merupakan negara yang memiliki "Discipline" paling populer di Olimpiade Tokyo 2020 dengan peserta berjumlah 5 atlet, yaitu dari Jamaika, Amerika Serikat, Belanda, Italia, Inggris dan Jerman.

###Apa saja jenis kelamin diseluruh "Discipline"?

```
Disc = genders.groupby('Discipline')
x = [Discipline for Discipline, df in Disc]
female = genders.Female
male = genders.Male
plt.figure(figsize=(20,8))
X_axis = np.arange(len(x))
plt.bar(X_axis - 0.2, female, 0.4, label = 'Female')
plt.bar(X_axis + 0.2, male, 0.4, label = 'Male')
plt.xlabel("Discipline",size=10)
plt.ylabel("Number of Participants")
plt.xticks(X_axis,x,rotation='vertical')
plt.legend()
plt.show()
```

Dari diagram diatas, hanya ada satu "Discipline" dengan lebih banyak wanita daripada pria yaitu Bersepeda BMX Freestyle. Namun, wanita berpartisipasi di dalam semua "Discipline" sedangkan pria tidak berpartisipasi di dalam semua "Discipline" seperti senam berirama dan berenang artistik.

###Negara mana yang paling banyak menerima medali emas? Mana yang paling banyak meraih perak dan perunggu? Mana yang menerima paling sedikit untuk masing-masing?

```
medals.rename(columns={'Team/NOC':'Team'},inplace=True) 
# Mengganti nama kolom Team/NOC ke Team
x = []
for team in medals.Team:
    x.append(team)
y = medals.Gold
plt.figure(figsize=(30,8))
plt.bar(x,y)
for index, value in enumerate(y):
    plt.text(index, value, str(value),color='blue',size=10,fontweight='bold')
plt.xlabel("Country",size=10)
plt.ylabel("Gold Medals")
plt.xticks(x,rotation='vertical',size=15)
plt.title("Gold Medals across Nations")
plt.show()
```

Dari hasil diagram diatas, Amerika Serikat memiliki medali emas terbanyak di antara semua negara yaitu 39 medali, lalu China memiliki selisih satu medali emas dari Amerika Serikat yaitu 38 medali dan Jepang memiliki 29 medali.

```
medals.rename(columns={'Team/NOC':'Team'},inplace=True) 
# Mengganti nama kolom Team/NOC column ke Team
x = []
for team in medals.Team:
    x.append(team)
y = medals.Silver
plt.figure(figsize=(30,8))
plt.bar(x,y)
for index, value in enumerate(y):
    plt.text(index, value, str(value),color='blue',size=10,fontweight='bold')
plt.xlabel("Country",size=10)
plt.ylabel("Silver Medals")
plt.xticks(x,rotation='vertical',size=15)
plt.title("Silver Medals across Nations")
plt.show()
```

Dari hasil diagram diatas, Amerika Serikat juga memiliki medali perak terbanyak di olimpiade ini dengan jumlah 41 medali dan China berada diurutan tertinggi kedua dengan jumlah 32 medali.

```
medals.rename(columns={'Team/NOC':'Team'},inplace=True) 
# Mengganti nama kolom Team/NOC ke Team
x = []
for team in medals.Team:
    x.append(team)
y = medals.Bronze
plt.figure(figsize=(30,8))
plt.bar(x,y)
for index, value in enumerate(y):
    plt.text(index, value, str(value),color='blue',size=10,fontweight='bold')
plt.xlabel("Country",size=10)
plt.ylabel("Bronze Medals")
plt.xticks(x,rotation='vertical',size=15)
plt.title("Bronze Medals across Nations")
plt.show()
```

Dari hasil diagram diatas, sekali lagi Amerika Serikat memiliki medali perunggu terbanyak dengan jumlah 33 medali. Kemudian, Rusia menduduki urutan tertinggi kedua dengan memiliki jumlah 23 medali.

##Final Analysis

1.   Sebagian besar althletes di Olimpiade berasal dari negara-negara seperti Amerika Serikat, Jepang dan Australia. Ini bisa jadi karena negara-negara ini mempromosikan olahraga dan atletik dari usia yang lebih muda, mengasah kaum muda dan mempersiapkan mereka untuk bersaing di tingkat yang lebih tinggi dan berinvestasi lebih banyak di belakang ini. Tidak seperti negara-negara seperti Tanzania atau Sudan, negara-negara ini mungkin memiliki budaya olahraga yang lebih baik.
2.   Atletik adalah "Discipline" paling populer di Olimpiade. Ini bisa jadi karena olahraga lain seperti sepak bola / bola basket mendapatkan panggung lain bagi pemain untuk menampilkan kelebihan mereka jadi mungkin inilah mengapa athletes yang tampil di atletik mencoba menampilkan kemampuan mereka di panggung besar ini.
3.   Jepang menghasilkan pelatih terbanyak dan AS setelahnya. Sekali lagi ini bisa jadi karena budaya olahraga yang luas di negara-negara ini dan ketika atlet tumbuh mempersiapkan diri dari usia yang jauh lebih muda, mereka mendapatkan banyak pengalaman, sehingga menjadi pelatih.
4.   Perempuan berpartisipasi dalam semua disiplin ilmu tetapi relatif kurang dari laki-laki. Ini tidak bisa karena segala jenis diskriminasi, melainkan kurangnya pertumbuhan pola pikir pada wanita muda untuk menjadi atlet hebat suatu hari nanti dibandingkan dengan pria. Tapi, tentunya statistik ini telah berkembang pesat dalam beberapa tahun terakhir dan suatu hari, wanita mungkin lebih mendominasi pada tahap-tahap ini juga.
5.   Amerika Serikat memiliki jumlah medali emas, perak dan perunggu tertinggi dan ini menunjukkan betapa suksesnya AS. Alasan terbesar bisa jadi kekayaan; lebih banyak investasi pada atlet dan disiplin masing-masing. Memiliki pelatih dan fasilitas yang lebih baik memang membantu menciptakan atlet yang lebih baik dan lebih baik sehingga kekayaan mungkin bisa menjadi alasan mengapa AS melakukan jauh lebih baik daripada negara lain.
